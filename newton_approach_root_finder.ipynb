```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import sympy as sy
from sympy import *
```

```python
def calculate_function(function ,derivate_function, x0):
    x1 = x0 - np.divide(function.subs(x,x0).evalf(20) , derivate_function.subs(x,x0).evalf(20))
    return x1
```

```python
from sympy import *
x = symbols("x")
```

```python
function = x + cos(x)
derivate_function = diff(function)
derivate_function
```

```python
num_list = []
x0 = 0
num_list.append(x0)
iterations = 30
for i in range(iterations):
    x_n = calculate_function(function=function , derivate_function=derivate_function ,x0=num_list[i])
    num_list.append(x_n)
    print('The result of iteration {} is {}'.format(i,x_n))
```
