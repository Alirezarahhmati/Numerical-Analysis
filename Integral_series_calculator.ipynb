```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import sympy as sy
from sympy import *
```

```python
from sympy import *
x = symbols("x")
```

```python
def calculate_bars(a:int,b:int,n:int,h:int,percision:int)->list:
    num_to_cal = np.linspace(0,1,n+1)
    bar_results = []
    for i in range(n+1):
        f_x = function.subs(x,num_to_cal[i]).evalf(percision)
        print("The evalf for substite of {} in function is {}".format(num_to_cal[i],f_x))
        bar_results.append(f_x)
    return bar_results    
```

```python
def calculate_integral(bar_results:list,h:int)->int:
    # result = np.add(np.multiply(np.divide(h,2),temp),result) 
    result = 0.0000
    for i in range(len(bar_results)):
        if(i==0 or i==(len(bar_results)-1)):
            result = np.add(bar_results[i],result) 
        result = np.add(np.multiply(2,bar_results[i]),result) 
    return np.multiply(np.divide(h,2),result)
```

```python
function = exp(-pow(x,2))
# 0-1

a = 0
b = 1 
n = 5000
h = (b-a)/n
percision = 30
```

```python
function
```

```python
bar_results = calculate_bars(a=a,b=b,n=n,h=h,percision=percision)
```

```python
result = calculate_integral(bar_results=bar_results,h=h)
print("The result for {} n is {}".format(n,result))
correct_answer = function.integrate(x).subs(x,1).evalf(percision)-function.integrate(x).subs(x,0).evalf(percision)
print('The correct answer is {}'.format(correct_answer))
print('The diffrence between the correct and predicted anser is {}'.format(np.subtract(result,correct_answer)))
```
